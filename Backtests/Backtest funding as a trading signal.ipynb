### 4. What happens if we trade only if the funding rate is bigger than a certain parameter?
When the funding rate is very small, opening a trade could lead only to increase the risk of the strategy due to the low expected reward.

The subset for Bitcoin is defined by a funding rate greater than abs(0.04%) while for Ethereum it's abs(0.09%). With these parameters, roughly 30% of the datapoints are analyzed.

```python
summary, trades = Strategy("XBTUSD", my_signal=["fundingRate"], exchanger="bitmex", short_param=[0.04/100], long_param=[-0.04/100], 
                            market_making=False, commission_market_taking=0,
                            window_open_wrt_signal=-240, window_mm_open=240, window_hold=24*60, window_mm_close=120 )
Plot_results(trades,asset = "XBTUSD",show_cumulative=True)
summary
```

### Ethereum

```python
summary, trades = Strategy("ETHUSD", my_signal=["fundingRate"], exchanger="bitmex", short_param=[0.09/100], long_param=[-0.09/100], 
                          market_making=False, commission_market_taking=0,
                          window_open_wrt_signal=-240, window_mm_open=240, window_hold=24*60, window_mm_close=120 )
Plot_results(trades,asset = "ETHUSD",show_cumulative=True)
summary
```
