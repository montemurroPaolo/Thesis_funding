#### Variables to run all the regressions

```python
import pandas as pd
filename   = "my_returns_daily.csv"
column     = "return_from_trade" #fundingRate or return_from_trade
```

#### Load, clean and merge data

```python
data_ff = pd.read_csv("ff.CSV")
data_ff.index = data_ff["date"]
```

```python
my_returns = pd.read_csv(filename) # insert my returns
```

```python
my_returns["date"] = [int(x.replace('-', '')) for x in my_returns["date"] ]
my_returns = my_returns.set_index("date", drop=True)
my_returns["my_returns"] = my_returns[column]*100
my_returns = my_returns[["my_returns"]]
```

```python
data = pd.merge(data_ff, my_returns["my_returns"],right_index = True, left_index=True,how="left")
data['my_returns'] = data['my_returns'].fillna(0)
data = data[data.date > my_returns.index.min() ]
```

```python
import statsmodels.formula.api as smf 
ff = smf.ols(formula='my_returns ~ mkt_excess + SMB + HML + RMW + CMA', data=data)
res = ff.fit()
res.summary()
```

### Hedge Fund Benchmarks: A Risk-Based Approach regression

#### Trend-Following Risk Factors (3):
- Bond Trend-Following Factor
- Currency Trend-Following Factor
- Commodity Trend-Following Factor

https://faculty.fuqua.duke.edu/~dah7/HFRFData.htm

```python
first3 = pd.read_csv("first_3_factors.csv")
first3 = first3.iloc[:,:-2]
first3["PTFSBD"]  = [float(first3["PTFSBD"][i].replace('%', ''))/100 for i in range(len(first3["PTFSBD"]))]
first3["PTFSFX"]  = [float(first3["PTFSFX"][i].replace('%', ''))/100 for i in range(len(first3["PTFSFX"]))]
first3["PTFSCOM"] = [float(first3["PTFSCOM"][i].replace('%', ''))/100 for i in range(len(first3["PTFSCOM"]))]
first3.columns  = ["Date","Bond_TF_factor", "Currency_TF_factor","Commodity_TF_factor"]
```

####  Equity Market Factor 
Standard & Poors 500 stock return. https://finance.yahoo.com/quote/%5EGSPC/history?period1=1437350400&period2=1595203200&interval=1mo&filter=history&frequency=1mo

```python
def Ret(df,col): return (  df[col] / df[col].shift(1) -1 )

gspc = pd.read_csv("^GSPC.csv")
gspc["Equity_market_factor"] = Ret(gspc, "Close")
gspc["Date"] = [float(gspc["Date"][i][:-3].replace('-', '')) for i in range(len(gspc["Date"]))]
```

#### The Size Spread Factor
Russell 2000 index monthly total return - Standard & Poors 500 monthly total return. https://finance.yahoo.com/quote/%5ERUT/history?period1=1437350400&period2=1595203200&interval=1d&filter=history&frequency=1d

```python
rut = pd.read_csv("^RUT.csv")
rut["ret"] = Ret(rut, "Close")
rut["Date"] = [float(rut["Date"][i][:-3].replace('-', '')) for i in range(len(rut["Date"]))]

rut["Size_spread_factor"] = rut["ret"] - gspc["Equity_market_factor"]
```

#### The Bond Market Factor - DGS10 (calculate change)
https://fred.stlouisfed.org/series/DGS10 

```python
bond_market_factor = pd.read_csv("DGS10.csv")
bond_market_factor["Date"] = [float(bond_market_factor["DATE"][i][:-3].replace('-', '')) for i in range(len(bond_market_factor["DATE"]))]
bond_market_factor = bond_market_factor[bond_market_factor["DGS10"] != "."]
bond_market_factor["DGS10"] = bond_market_factor["DGS10"].astype(float)
bond_market_factor = bond_market_factor[bond_market_factor.groupby('Date').cumcount() == 0] # https://stackoverflow.com/questions/21591326/python-pandas-select-first-observation-per-group
bond_market_factor["Bond_market_factor"] = Ret(bond_market_factor, "DGS10")
```

#### The Credit Spread Factor DGS10/DBA
 The monthly change in the Moody's Baa yield less 10-year treasury constant maturity yield (month end-to-month end)
https://fred.stlouisfed.org/series/DBAA

```python
credit_spread_factor = pd.read_csv("DBAA.csv")
credit_spread_factor["Date"] = [float(credit_spread_factor["DATE"][i][:-3].replace('-', '')) for i in range(len(credit_spread_factor["DATE"]))]
credit_spread_factor = credit_spread_factor[credit_spread_factor["DBAA"] != "."]
credit_spread_factor["DBAA"] = credit_spread_factor["DBAA"].astype(float)
credit_spread_factor = credit_spread_factor[credit_spread_factor.groupby('Date').cumcount() == 0] # https://stackoverflow.com/questions/21591326/python-pandas-select-first-observation-per-group
credit_spread_factor["change"] = Ret(credit_spread_factor, "DBAA")
credit_spread_factor["Credit_spread_factor"] = credit_spread_factor["change"] - bond_market_factor["Bond_market_factor"]
```

#### The Emerging Market Index EEM
MSCI Emerging Market index monthly total return https://it.finance.yahoo.com/quote/EEM/history?period1=1437350400&period2=1595203200&interval=1d&filter=history&frequency=1d

```python
eem = pd.read_csv("EEM.csv")
eem["Emergency_market_factor"] = Ret(eem, "Close")
eem["Date"] = [float(eem["Date"][i][:-3].replace('-', '')) for i in range(len(eem["Date"]))]
```

Creating the factors dataset

```python
factors8 = pd.merge(first3, gspc[["Date","Equity_market_factor"]], how="left", on="Date")
factors8 = pd.merge(factors8, rut[["Date","Size_spread_factor"]], how="left", on="Date")
factors8 = pd.merge(factors8, bond_market_factor[["Date","Bond_market_factor"]], how="left", on="Date")
factors8 = pd.merge(factors8, credit_spread_factor[["Date","Credit_spread_factor"]], how="left", on="Date")
factors8 = pd.merge(factors8, eem[["Date","Emergency_market_factor"]], how="left", on="Date")

factors8 = factors8.set_index("Date")
factors8 = factors8[factors8.index>201508]
factors8 = factors8 * 100
```

```python
#### Our returns as monthly returns and merge
my_returns["Date"] = [int(str(my_returns.index[i])[:-2]) for i in range(len(my_returns))]
my_returns = my_returns.groupby("Date").sum()["my_returns"]

factors8 = pd.merge(my_returns, factors8, how="left", on="Date")
```

```python
# Regression
ff = smf.ols(formula='my_returns ~ Bond_TF_factor + Currency_TF_factor + Commodity_TF_factor + Equity_market_factor + Size_spread_factor + Bond_market_factor + Credit_spread_factor + Emergency_market_factor', data=factors8)
res = ff.fit()
res.summary()
```

## Regress over crypto variables

```python
eth_binance = pd.read_csv("/Users/paolo/Desktop/Thesis_code/data/ETHUSD-1d-binance-data-spot.csv")
eth_bitmex = pd.read_csv("/Users/paolo/Desktop/Thesis_code/data/ETHUSD-1d-bitmex-data-future.csv")
xbt_binance = pd.read_csv("/Users/paolo/Desktop/Thesis_code/data/XBTUSD-1d-binance-data-spot.csv")
xbt_bitmex = pd.read_csv("/Users/paolo/Desktop/Thesis_code/data/XBTUSD-1d-bitmex-data-future.csv")

# Unify dates
eth_bitmex["timestamp"] = [eth_bitmex["timestamp"][i][:-15] for i in range(len(eth_bitmex["timestamp"]))]
xbt_bitmex["timestamp"] = [xbt_bitmex["timestamp"][i][:-15] for i in range(len(xbt_bitmex["timestamp"]))]
```

```python
# merge prices
eth = pd.merge(eth_bitmex[["timestamp","close","open"]], eth_binance[["timestamp","open"]], how="left", on="timestamp")
xbt = pd.merge(xbt_bitmex[["timestamp","close","open"]], xbt_binance[["timestamp","open"]], how="left", on="timestamp")
```

```python
#Create premium (future bitmex vs spot binance)
eth["premium_eth"] = eth["close"] / eth["open_y"] -1
xbt["premium_xbt"] = xbt["close"] / xbt["open_y"] -1

# Create returns
xbt["return_xbt"] = xbt["close"] / xbt["open_x"] -1
eth["return_eth"] = eth["close"] / eth["open_x"] -1 

# Rename
eth.columns  = ["timestamp","close_eth","open_eth","spot_eth","premium_eth","return_eth"]
xbt.columns  = ["timestamp","close_xbt","open_xbt","spot_xbt","premium_xbt","return_xbt"]

# Merge
data = pd.merge(xbt, eth, how="left", on="timestamp")
```

```python
# Load funding data on eth
funding_eth = pd.read_csv("/Users/paolo/Desktop/Thesis_code/data/ETHUSD-bitmex-funding-data.csv")
funding_eth["timestamp"] = [funding_eth["timestamp"][i][:-15] for i in range(len(funding_eth["timestamp"]))]
funding_eth = funding_eth.groupby('timestamp').first()

# Merge and clean
data = pd.merge(data, funding_eth[["fundingRate"]], on="timestamp")
data["timestamp"] = [x.replace('-', '').replace('', '') for x in data["timestamp"] ]
data["timestamp"] = data["timestamp"].astype(int)
```

```python
# Unify and merge
my_returns = my_returns/100 #(before it was another scale..)
data = pd.merge(my_returns, data, right_on="timestamp",left_on="date", how="left")
```

```python
# Regression
ff = smf.ols(formula='my_returns ~ fundingRate + premium_xbt + premium_eth + return_xbt + return_eth', data=data)
res = ff.fit()
res.summary()
```
